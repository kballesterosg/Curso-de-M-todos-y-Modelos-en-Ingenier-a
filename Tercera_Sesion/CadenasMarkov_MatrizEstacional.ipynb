# Vector Estacional en Cadenas de Markov

## ¿Qué es una cadena de Markov?

Una cadena de Markov es un modelo probabilístico que describe un sistema que **evoluciona entre estados** con **transiciones que dependen solo del estado actual** (no del pasado).

---

## ¿Qué es la matriz o vector estacional?

En una cadena de Markov, si dejamos que el sistema evolucione durante muchos pasos, las probabilidades de estar en cada estado tienden a estabilizarse. Ese comportamiento a largo plazo se describe mediante el:

### Vector Estacional $( \pi )$

Es un vector de probabilidades tal que:

$
\pi P = \pi
$

con la condición:

$
\sum_i \pi_i = 1
$

**Interpretación:**  
El vector \( \pi \) indica la **distribución de probabilidad estable** de los estados a largo plazo. Es decir, la proporción del tiempo que el sistema pasa en cada estado, sin importar el estado inicial.

---

## Ejemplo:

Supón la siguiente matriz de transición \( P \):

$
P =
\begin{bmatrix}
0.6 & 0.3 & 0.1 \\
0.2 & 0.5 & 0.3 \\
0.1 & 0.2 & 0.7
\end{bmatrix}
$

Queremos encontrar el vector estacional $( \pi = [\pi_1, \pi_2] )$ tal que:

$
[\pi_1, \pi_2] \cdot P = [\pi_1, \pi_2]
\quad \text{y} \quad \pi_1 + \pi_2 = 1
$

---

## Propiedades clave del vector estacional

- Existe si la cadena es **irreducible** (todos los estados se comunican entre sí) y **aperiódica**.
- Es **único** bajo esas condiciones.
- Todas las **filas de $( P^n )$** tienden a parecerse a medida que $( n \to \infty )$, convergiendo al mismo vector estacional.

---

## En resumen

| Concepto              | Significado                                              |
|-----------------------|----------------------------------------------------------|
| **Vector estacional** | Vector $( \pi )$ tal que $( \pi P = \pi )$ y $( \sum \pi_i = 1 )$ |
| **Interpretación**    | Distribución a largo plazo de los estados                |
| **Importancia**       | Describe el equilibrio del sistema a largo plazo         |

# **Cálculo Automatico de la Matriz o Vector Estacional**



```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transición
P = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.5, 0.3],
    [0.1, 0.2, 0.7]
])

# Inicialización
n_steps = 20
powers = [P.copy()]
for _ in range(1, n_steps + 1):
    powers.append(powers[-1] @ P)

# Mostrar algunas potencias redondeadas
for i, matrix in enumerate(powers[1:], 1):
    print(f"P^{i} =\n", np.round(matrix, 3), "\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transición
P = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.5, 0.3],
    [0.1, 0.2, 0.7]
])

# Inicialización
n_steps = 12
powers = [P.copy()]
for _ in range(1, n_steps + 1):
    powers.append(powers[-1] @ P)

# Mostrar algunas potencias redondeadas
for i, matrix in enumerate(powers[1:], 1):
    print(f"P{i} =\n", np.round(matrix, 3), "\n")

# Graficar evolución de las filas de la matriz
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

for row in range(3):
    for col in range(3):
        axs[row].plot([P_k[row, col] for P_k in powers], label=f"Estado {col+1}")
    axs[row].set_title(f"Evolución de la fila {row+1}")
    axs[row].set_ylabel("Probabilidad")
    axs[row].legend()
    axs[row].grid(True)

plt.xlabel("Potencia de la matriz P (n)")
plt.suptitle("Evolución de la matriz de transición P^n")
plt.tight_layout()
plt.show()
```

# Simulación de la evolución en una cadena de Markov

## Objetivo
Modelar cómo cambian las preferencias de una población entre tres tipos de gaseosas (A, B y C) a lo largo del tiempo, usando una **cadena de Markov**.

## Datos iniciales

- La población inicial es:
  - 200 personas prefieren gaseosa A
  - 300 personas prefieren gaseosa B
  - 500 personas prefieren gaseosa C

- La **matriz de transición** $( P )$ indica las probabilidades de cambiar de una preferencia a otra:

$
P = \begin{bmatrix}
0.6 & 0.3 & 0.1 \\
0.2 & 0.5 & 0.3 \\
0.1 & 0.2 & 0.7
\end{bmatrix}
$

## Simulación

Se simulan 12 pasos (por ejemplo, 12 meses), multiplicando en cada paso el vector de estado actual por la matriz de transición:

$
v_{t+1} = v_t \cdot P
$

## Gráfico

El gráfico muestra cómo cambia la cantidad de personas que prefieren cada tipo de gaseosa a lo largo del tiempo. Es una forma visual de observar cómo el sistema evoluciona hacia un equilibrio o estado estacionario.



```python
import numpy as np

# Matriz de transición
P = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.5, 0.3],
    [0.1, 0.2, 0.7]
])

# Vector de estado inicial (población que prefiere cada tipo de gaseosa)
v0 = np.array([200, 300, 500])


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transición
P = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.5, 0.3],
    [0.1, 0.2, 0.7]
])

# Vector de estado inicial (población que prefiere cada tipo de gaseosa)
v0 = np.array([200, 300, 500])

# Número de pasos (por ejemplo, 12 meses)
n_pasos = 20

# Almacenar la evolución
estados = [v0]

# Simulación paso a paso
for _ in range(n_pasos):
    v_siguiente = estados[-1] @ P
    estados.append(v_siguiente)

print(np.round(estados[-1], 0))

# Convertimos la lista a un array para graficar
estados = np.array(estados)

# Graficar la evolución
plt.figure(figsize=(10, 6))
plt.plot(estados[:, 0], label='Gaseosa A')
plt.plot(estados[:, 1], label='Gaseosa B')
plt.plot(estados[:, 2], label='Gaseosa C')
plt.xlabel('Paso del tiempo (meses)')
plt.ylabel('Población')
plt.title('Evolución de preferencias por gaseosa')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

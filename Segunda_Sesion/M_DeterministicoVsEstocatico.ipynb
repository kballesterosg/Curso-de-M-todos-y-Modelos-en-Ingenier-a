### Comparación de modelos: crecimiento poblacional

---

El modelo de crecimiento poblacional es un ejemplo clásico para ilustrar la diferencia entre enfoques determinísticos y estocásticos en la modelación matemática. En el modelo determinístico, se asume que la población crece de forma continua y predecible siguiendo una función exponencial, donde la tasa de crecimiento es constante. Esta formulación implica que, dadas unas condiciones iniciales, el comportamiento del sistema será siempre el mismo. En contraste, el modelo estocástico introduce una componente aleatoria en la tasa de crecimiento, representando la variabilidad natural del sistema debida a factores como migraciones, eventos climáticos o cambios en las tasas de natalidad. De este modo, cada simulación del modelo estocástico puede generar trayectorias distintas, proporcionando una representación más realista de la incertidumbre en los procesos poblacionales. Esta comparación permite evidenciar cómo la inclusión de la aleatoriedad puede enriquecer el análisis y la interpretación de fenómenos dinámicos en contextos reales.

#### **1. Modelo Determinístico**

Usamos el clásico **modelo exponencial**:

$P(t) = P_0 \cdot e^{r t}$

- $( P_0 )$: población inicial  
- $( r )$: tasa de crecimiento  
- $( t )$: tiempo

Este modelo **siempre da el mismo resultado** si usamos los mismos parámetros.

---

#### **2. Modelo Estocástico**

Agregamos una **componente aleatoria** a la tasa de crecimiento:

$P(t+\Delta t) = P(t) \cdot e^{(r + \epsilon_t)\Delta t}$

Donde:


- $(εₜ ∼ N(0, σ))$: es un **ruido aleatorio** tomado de una distribución normal con media 0 y desviación estándar $(\sigma)$

Cada simulación puede dar un **resultado diferente**, como ocurre en la vida real con nacimientos, migraciones, enfermedades, etc.

### Ejemplos reales de lo que puede representar $( \epsilon_t )$ en el crecimiento poblacional

El término $( \epsilon_t )$ representa eventos aleatorios o perturbaciones que afectan la tasa de crecimiento en un momento dado. Algunos ejemplos concretos podrían ser:

- **Un año con brote de enfermedad**  
  $( \epsilon_t < 0 )$ → la tasa de crecimiento disminuye, puede incluso haber una reducción en la población

- **Un año con programas de salud exitosos o migración positiva**  
  $( \epsilon_t > 0 )$ → la tasa de crecimiento se incrementa por encima del valor esperado

- **Desastres naturales o conflictos sociales**  
  Afectan negativamente la tasa de crecimiento (posiblemente $( \epsilon_t < 0 )$)

- **Cambios económicos, ambientales o sociales inesperados**  
  Introducen variabilidad en el comportamiento poblacional, positiva o negativa según el evento

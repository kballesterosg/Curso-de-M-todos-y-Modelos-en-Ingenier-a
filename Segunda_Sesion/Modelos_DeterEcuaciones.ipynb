## Modelo de crecimiento de bacterias

Este modelo describe cómo crece una colonia de bacterias que **se duplica cada día**. Es un ejemplo clásico de crecimiento exponencial, donde el cambio en el número de bacterias es **proporcional a la cantidad que ya existe**.

### Ecuación diferencial

Partimos de la siguiente ecuación diferencial de primer orden:

$\frac{dy}{dt} = ky$

donde:
- $y(t)$ es el número de bacterias en el tiempo $( t )$,
- $k$ es la constante de crecimiento,
- $t$ es el tiempo medido en días.

---

### Solución general

La solución de esta ecuación es:

$y(t) = C_1 e^{kt}$

Usamos la condición inicial $y(0) = 100$ bacterias para encontrar $( C_1 )$:

$y(0) = C_1 e^{k \cdot 0} = C_1 \cdot 1 = 100 \Rightarrow C_1 = 100$

Entonces, la solución queda:

$y(t) = 100 e^{kt}$

---

### Cálculo de la constante $( k )$

Si después de 1 día hay 200 bacterias, usamos ese dato para encontrar $( k )$:

$y(1) = 100 e^k = 200 \Rightarrow e^k = 2 \Rightarrow k = \ln(2)$

---

### Modelo final

Sustituyendo $( k = \ln(2) )$, el modelo queda:

$y(t) = 100 \cdot e^{\ln(2)t} = 100 \cdot 2^t$

Este modelo muestra que el número de bacterias **se duplica cada día**, como se esperaba.

---

### Interpretación

- Día 0 → 100 bacterias  
- Día 1 → 200 bacterias  
- Día 2 → 400 bacterias  
- Día 3 → 800 bacterias  
- Día 4 → 1600 bacterias  

El crecimiento es exponencial, porque en cada unidad de tiempo, la cantidad se **multiplica** por 2.

### Modelo Hidrológico Simple: Balance Lluvia–Escorrentía

---

#### Ecuación del modelo:

$
\frac{dS}{dt} = P(t) - Q(t) - E(t)
$

Donde:

- $S(t)$: almacenamiento de agua en el suelo o cuenca (mm)
- $P(t)$: precipitación (mm/h)
- $Q(t)$: escorrentía superficial (mm/h)
- $E(t)$: evapotranspiración (mm/h)

---

#### Supuestos del ejercicio:

- La lluvia es constante durante las primeras 6 horas: $( P(t) = 5 \text{mm/h} )$
- La escorrentía es proporcional al almacenamiento: $( Q(t) = c \cdot S(t) )$ con $( c = 0.05 )$.
- La evapotranspiración es constante: $( E(t) = 0.2 \, \text{mm/h} )$
- El paso de tiempo es 1 hora
- Se simula durante 24 horas

---

### Descripción del ejercicio:

En este ejercicio se simula el comportamiento del almacenamiento de agua en una cuenca o suelo ante un evento de lluvia. El modelo es determinista y se basa en un balance hídrico simple que considera:

- Entrada por precipitación
- Salida por escorrentía (dependiente del almacenamiento)
- Pérdidas por evapotranspiración

Durante las primeras horas (0 a 6 h), el sistema recibe lluvia constante, lo que hace que el almacenamiento aumente. Una vez cesa la lluvia, el almacenamiento disminuye progresivamente debido a la escorrentía y la evapotranspiración.

---

### ¿Por qué se llama "Solución numérica con Euler"?

El modelo hidrológico simple que usamos está basado en una **ecuación diferencial** del tipo:

$
\frac{dS}{dt} = P - cS - E
$

Donde la derivada nos indica cómo cambia el almacenamiento de agua $( S )$ con respecto al tiempo.

---

### ¿Qué es el método de Euler?

Es un **método numérico** sencillo que se usa para aproximar soluciones de ecuaciones diferenciales cuando:

- No se tiene una solución analítica cerrada
- Se desea simular el comportamiento paso a paso en el tiempo

Euler aproxima la curva de solución usando la **pendiente local** en cada punto:

$
S_{n+1} = S_n + \left(P - c S_n - E \right) \cdot \Delta t
$

Es decir, el nuevo valor de $( S )$ se calcula como el anterior más un incremento que depende de la derivada en ese punto.

---

### ¿Cómo se aplica?

1. Se define una condición inicial $( S_0 )$
2. Se escoge un paso de tiempo $( \Delta t )$
3. Se aplica la fórmula de Euler en cada paso para construir la trayectoria de $( S )$

---

# Modelo de Crecimiento de Bosque y Captura de Carbono

Este modelo representa el crecimiento de un bosque y la captura de carbono asociada mediante un enfoque matemático sencillo pero útil para análisis exploratorios. El crecimiento de la biomasa forestal se modela con una ecuación logística, que considera una tasa de crecimiento intrínseca limitada por una capacidad máxima del ecosistema (biomasa máxima que puede sostener). Además, se incluye una tasa de deforestación que reduce la biomasa disponible en función del tiempo. La captura de carbono, por su parte, se asume proporcional al incremento neto de biomasa, considerando un factor de conversión que refleja la cantidad de dióxido de carbono almacenado por tonelada de biomasa generada. Este tipo de modelo es útil para analizar escenarios de conservación, restauración ecológica o impacto de políticas de uso del suelo en la mitigación del cambio climático.


## Variables:
- $B(t) $: Biomasa forestal (toneladas/hectárea) en el tiempo \( t \)
- $ C(t)$: Carbono capturado acumulado (toneladas de CO₂/hectárea)
- $ r $: Tasa de crecimiento intrínseca del bosque
- $ K $: Capacidad máxima de biomasa (toneladas/hectárea)
- $ \alpha $: Factor de captura de carbono (tCO₂/t biomasa)
- $ d $: Tasa de deforestación (proporción de biomasa perdida por año)

---

## 1. Ecuación del crecimiento logístico con deforestación

$
\frac{dB}{dt} = r B \left(1 - \frac{B}{K} \right) - d B
$

- El primer término representa el crecimiento natural del bosque.
- El segundo término representa la pérdida por deforestación.

---

## 2. Ecuación de captura de carbono

$
\frac{dC}{dt} = \alpha \cdot \frac{dB}{dt}
$

- El carbono capturado es proporcional al cambio en biomasa.

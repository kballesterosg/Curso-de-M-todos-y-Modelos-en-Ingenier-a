### Distribución Normal

La distribución normal (o gaussiana) es una distribución de probabilidad continua, simétrica respecto a su media. Es común en fenómenos naturales y en errores de medición. La distribución normal representa fenómenos naturales como la altura o el peso.

**Parámetros:**

- **μ (mu)**: media
- **σ (sigma)**: desviación estándar


**Características:**
- Forma de campana.
- Simétrica.
- Dominio: todos los números reales.
- Aplicaciones: alturas, pesos, errores.

**Función de densidad:**
$
  f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$


```python
import numpy as np
import matplotlib.pyplot as plt
#****************
import seaborn as sns
sns.set(style="whitegrid")
#****************

mu = 100
sigma = 10
n = 1000

data = np.random.normal(mu, sigma, n)

plt.hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
#sns.histplot(data, bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title("Distribución Normal (μ=100, σ=10)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Distribución Exponencial

Es una distribución continua, no simétrica. Modela el tiempo entre eventos.

**Parámetro:**
- **λ (lambda)**: tasa media de ocurrencia (la media es 1/λ)

**Características:**
- Sesgada hacia la derecha.
- Dominio: [0, ∞)
- Aplicaciones: tiempo entre fallos, tiempo de espera.

**Función de densidad:**
$
  f(x) = \lambda e^{-\lambda x},\ x \geq 0
$

```python
import numpy as np
import matplotlib.pyplot as plt

scale = 1
n = 1000

data = np.random.exponential(scale, n)

plt.hist(data, bins=30, density=True, color='lightgreen', edgecolor='black')
plt.title("Distribución Exponencial (λ=1)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Distribución Uniforme

Todos los valores dentro de un intervalo tienen la misma probabilidad.

**Parámetros:**
- **a**: mínimo
- **b**: máximo

**Características:**
- Forma rectangular.
- Dominio: [a, b]
- Aplicaciones: simulaciones, distribución aleatoria sin sesgo.

**Función de densidad:**
$
  f(x) = \frac{1}{b-a},\ a \leq x \leq b
$

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 1
n = 1000

data = np.random.uniform(a, b, n)

plt.hist(data, bins=30, density=True, color='orange', edgecolor='black')
plt.title("Distribución Uniforme (0, 1)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Distribución Poisson

Modelo de conteo de eventos que ocurren en intervalos fijos de tiempo o espacio. Se usa para contar eventos que ocurren de forma aleatoria, pero con una tasa promedio constante.

**Parámetro:**
- **λ (lambda)**: número medio de eventos en el intervalo

**Características:**
- Discreta (0, 1, 2, ...)
- Asimétrica si λ es pequeño.
- Aplicaciones: llamadas por hora, accidentes por día.

**Función de probabilidad:**
$
  P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$

```python
import numpy as np
import matplotlib.pyplot as plt

lam = 5
n = 1000

data = np.random.poisson(lam, n)

plt.hist(data, bins=range(0, max(data)+1), density=True, color='purple', edgecolor='black', align='left')
plt.title("Distribución Poisson (λ=5)")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

### Distribución Weibull

Modela tiempos de falla o vida útil de productos. Usada para modelar tiempos de vida o fallos de sistemas.

- **a**: forma (shape)

**Características:**
- Puede ser creciente o decreciente dependiendo de la forma.
- Dominio: [0, ∞)
- Aplicaciones: análisis de confiabilidad, duración de productos.

**Función de densidad:**
$
  f(x) = \frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} e^{-(x/\lambda)^k},\ x \geq 0
$

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1.5
n = 1000

data = np.random.weibull(a, n)

plt.hist(data, bins=30, density=True, color='pink', edgecolor='black')
plt.title("Distribución Weibull (a=1.5)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Distribución Beta

Modela proporciones y probabilidades. Modelo flexible de variables continuas en [0, 1].

**Parámetros:**
- **α (alpha)**: forma
- **β (beta)**: forma


**Características:**
- Puede tomar muchas formas según α y β.
- Dominio: [0, 1]
- Aplicaciones: proporciones, probabilidades subjetivas.

**Función de densidad:**
$
  f(x) = \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}}{B(\alpha, \beta)}
$

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 2
beta = 5
n = 1000

data = np.random.beta(alpha, beta, n)

plt.hist(data, bins=30, density=True, color='gold', edgecolor='black')
plt.title("Distribución Beta (α=2, β=5)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Distribución Triangular

Modelo de distribución continua basada en un valor mínimo, máximo y modo. Útil cuando solo se conoce el valor mínimo, el más probable y el máximo.

**Parámetros:**
- **a**: mínimo
- **b**: máximo
- **c**: valor más probable (moda)

**Características:**
- Forma de triángulo.
- Dominio: [a, b]
- Aplicaciones: estimaciones subjetivas en gestión de proyectos.

**Función de densidad:**
$
  f(x) = \begin{cases}
    0 & x < a \\
    \frac{2(x-a)}{(b-a)(c-a)} & a \leq x < c \\
    \frac{2(b-x)}{(b-a)(b-c)} & c \leq x \leq b \\
    0 & x > b
  \end{cases}
$

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
c = 5
b = 10
n = 1000

data = np.random.triangular(a, c, b, n)

plt.hist(data, bins=30, density=True, color='gray', edgecolor='black')
plt.title("Distribución Triangular (0, 5, 10)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

# **Gráfico de todas las distribuciones con la Librería Seaborne**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tamaño de la muestra
n = 1000

# Crear distribuciones
distributions = {
    'Normal (μ=100, σ=10)': np.random.normal(loc=100, scale=10, size=n),
    'Exponencial (λ=1)': np.random.exponential(scale=1, size=n),
    'Uniforme (0, 1)': np.random.uniform(0, 1, size=n),
    'Poisson (λ=5)': np.random.poisson(lam=5, size=n),
    'Weibull (a=1.5)': np.random.weibull(a=1.5, size=n),
    'Beta (α=2, β=5)': np.random.beta(a=2, b=5, size=n),
    'Triangular (0, 5, 10)': np.random.triangular(left=0, mode=5, right=10, size=n)
}

# Graficar
plt.figure(figsize=(15, 12))
for i, (name, data) in enumerate(distributions.items(), 1):
    plt.subplot(4, 2, i)
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.title(name)
    plt.grid(True)

plt.tight_layout()
plt.show()
```
